```python
import pandas as pd
import numpy as np


import scipy.stats as sts
from scipy.optimize import curve_fit

import matplotlib.pyplot as plt
```

```python
url = "https://gist.githubusercontent.com/raquelhr/d1324510056353feeccf111d6b186a0d/raw/7b3bccc7917f3baa7ec1d919195d120083ee75e9/proctatinium_data.csv"

data = pd.read_csv(url)
t_values = np.array(data["time"], dtype=float)
count_values = np.array(data["count_rate"], dtype=float)
```

```python
plt.plot(t_values, count_values)
plt.ylabel("count rate")
plt.xlabel("time")
plt.show()
```

```python
N = 10000
t = np.linspace(0, np.max(t_values), 500)

# Initial lambda guess
Lambda = np.logspace(1e-5,1e5, N)
N_0 = 32

results = np.zeros(N)
for i in range(N):
    l = Lambda[i]
    cts = N_0*np.exp(-l*t) + np.random.normal(0, 1, size = len(t))

    results[i] = curve_fit(eN, t, cts, 10)[0]
```

```python
def eN(t, Lambda): 
    return 32*np.exp(-Lambda*t)

```
