## Sayama,  Exercise 11.3
Fit game of life. 

```python
# probability of initially panicky individuals

density = []
n =100

def initialize(p = 0.1 ):
    global config, nextconfig
    config = zeros([n, n])
    for x in range(n):
        for y in range(n):
            config[x, y] = 1 if random() < p else 0
    nextconfig = zeros([n, n])
    
def observe():
    global config, nextconfig
    cla()
    imshow(config, vmin = 0, vmax = 1, cmap = cm.binary)

def update():
    global config, nextconfig, density
    n_black = 0 
    for x in range(n):
        for y in range(n):
            count = 0
            for dx in [-1, 0, 1]:
                for dy in [-1, 0, 1]:
                    if dx == 0 and dy == 0:
                        count += config[(x + dx) % n, (y + dy) % n]
                        n_black += config[(x + dx) % n, (y + dy) % n]
                    else:
                        count += config[(x + dx) % n, (y + dy) % n]
            curr_config = config[x,y]
            # If alive (==1), stay alive if 2 or 3 are alive
            if curr_config:
                nextconfig[x, y] = 1 if count == 3 or count == 2 else 0
            else:
                nextconfig[x,y] = 1 if count == 3 else 0

    curr_density = n_black/(n*n)

    density.append(curr_density)

    config, nextconfig = nextconfig, config

import pycxsimulator

sim = pycxsimulator.GUI().start(func=[initialize, observe, update])
```

# Sayama, Exercise 11.5

Add p and update

```python

n = 100 # size of space: n x n
endpoint = []

class Simulation:

    def __init__(self, p = 0.1, steps = 5) -> None:
        self.steps = steps
        self.current_step = 0
        self.p = p
        self.config = []
        self.nextconfig = []
        self.density = []
        self.n = 100
    

    def run(self):
        self.initialize()
        while self.current_step < self.steps:
            self.update()

    def initialize(self):
        n = self.n
        config = zeros([n, n])

        for x in range(n):
            for y in range(n):
                config[x, y] = 1 if random() < self.p else 0
        
        self.config = config
        self.nextconfig = zeros([n, n])


    def update(self):
        n_black = 0
        for x in range(self.n):
            for y in range(self.n):
                count = 0
                for dx in [-1, 0, 1]:
                    for dy in [-1, 0, 1]:
                        count += self.config[(x + dx) % n, (y + dy) % n]
                self.nextconfig[x, y] = 1 if count >= 4 else 0
                n_black += self.config[x,y]
        
        self.config, self.nextconfig = self.nextconfig, self.config
        self.current_step += 1
        
        self.density.append(n_black/(n*n))    
```

```python
runs  = []

for prob in np.linspace(0,1,100):
    sim = Simulation(p = prob)
    sim.run()

    runs.append((prob, sim.density[-1]))

```
