```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as st
```

```python
def draw_gamma(alpha, beta):

    alpha  = alpha
    beta = beta

    assert len(alpha) == len(beta)

    x = np.linspace(0,15)

    for i in range(len(alpha)):
        dist = st.distributions.gamma(alpha[i], beta[i])
        plt.plot(x, dist.pdf(x), label = f'Gamma pdf for a={alpha[i]} b={beta[i]}')

    samples = dist.rvs(1000)
    plt.hist(samples, 
             label = f"100 sample gamma hist for a={alpha[1]} b={beta[1]}",
             alpha = 0.5,
             density= True,
             bins = 100)

    plt.legend()


alpha = [3,2]
beta = [3,4]

draw_gamma(alpha, beta)
```

```python
def draw_dist(dist, *args):
    args = args

    dist_mapper = {
        "beta" : st.distributions.beta,
        "gamma" : st.distributions.gamma,
        "normal" : st.distributions.norm
    }
    
    dist  = dist_mapper[dist]

    if len(args) > 1 and type(args[0]) == list:
        assert len(args[0]) == len(args[1]) 

        for plot in range(len(args[0])):
            x = np.linspace(0, 20)
            curr_args = [params[plot] for params in args]
            y = dist.pdf(x, *curr_args)
            plt.plot(x,y)
        return plt.show()
    
    x = np.linspace(-40, 40)
    y = dist.pdf(x, *args)

    return plt.plot(x,y)

draw_dist("normal", [4,6,7,5,7], [4,9,7,6,8])
```

```python
def plot_two(dist, *args, **kwargs):
    args = args
    kwargs = kwargs

    return args, kwargs

plot_two("dist", 10, 5, alpha = 5, beta = 10)
```
